## The DataFrame order matters

Visualize the following `DataFrame` with bars in decreasing order.

```python
df = (df_energy
    .query('year == 2021')
    .groupby('technology').generation_mwh.sum()
    .reset_index()
)

df
```

### Dummy

```python
import plotly.express as px
px.bar(data_frame=df, y='technology', x='generation_mwh')
```

### Proficient

```python
df = df.sort_values('generation_mwh')
df
```

```python
px.bar(data_frame=df, y='technology', x='generation_mwh')
```

Also, you may reverse the visualization order `Ascending=False`

```python
df = df.sort_values('generation_mwh', ascending=False)
df
```

```python
import plotly.express as px

px.bar(data_frame=df, y='technology', x='generation_mwh')
```

## Pivot tables vs groupby

Summarise the `generation_mwh` by `technology` and `month`.

```python
df = df_energy.query('year == 2021')
df
```

### Dummy

```python
(df
 .groupby(['technology', 'month'])
    .generation_mwh.sum()
 .reset_index()
)
```

### Proficient

```python
df.pivot_table(index='technology', columns='month', values='generation_mwh', aggfunc='sum')
```

## From absolute to the thousands

Don't show all digits of the thousands. For example: 20 000 should be 20.

### Dummy

```python
(df
 .style
    .background_gradient()
    .format(precision=2)
)
```

### Proficient

```python
(df
 .div(1000)
 .style
    .background_gradient()
    .format(precision=2)
)
```

## Specify the darkest value

### Dummy

```python
df = df.div(1000)

(df
 .style
    .background_gradient()
    .format(precision=2)
)
```

### Proficient

```python
(df
 .style
    .background_gradient(vmin=df.values.min(), vmax=df.values.max())
    .format(precision=2)
)
```
