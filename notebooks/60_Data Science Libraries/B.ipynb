# Python libraries for data science

## Numerical operations: `numpy`

https://numpy.org/doc/stable/reference/random/index.html

```python
import numpy as np
```

```python
salary_madrid = np.random.normal(25, 3, 100)
salary_barcelona = 1.2 * salary_madrid + np.random.normal(0, 2, 100)
```

## Graphs and plots: `plotly`

https://plotly.com/python/plotly-express/

```python
import plotly.express as px
```

### Line plot

```python
px.line(x=salary_madrid, y=salary_barcelona)
```

### Density contour plot

```python
px.density_contour(x=salary_madrid, y=salary_barcelona)
```

### Maps

```python
import plotly.express as px

df = px.data.election()
geojson = px.data.election_geojson()

fig = px.choropleth_mapbox(df, geojson=geojson, color="Bergeron",
                           locations="district", featureidkey="properties.district",
                           center={"lat": 45.5517, "lon": -73.7073},
                           mapbox_style="carto-positron", zoom=9)

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

### Scatter plot

```python
px.scatter(x=salary_madrid, y=salary_barcelona)
```

## Machine Learning: `sklearn`

https://scikit-learn.org/stable/user_guide.html

### Linear regression

```python
from sklearn.linear_model import LinearRegression

explanatory = df[['F']]
target = df['Pts']

model = LinearRegression()
model.fit(explanatory, target)
```

```python
df_pred = df[['F', 'Pts']].copy()
df_pred['Pts_pred_lr'] = model.predict(explanatory)
df_pred
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['F'], df['Pts'], label='Observed data')
plt.plot(df['F'], df_pred['Pts_pred_lr'], color='orange', label='Prediction: Linear Regression', marker='o')
plt.legend()
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

explanatory = df[['F']]
target = df['Pts']

model = DecisionTreeRegressor()
model.fit(explanatory, target)
```

```python
df_pred['Pts_pred_dt'] = model.predict(explanatory)
df_pred
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['F'], df['Pts'], label='Observed data')
plt.plot(df['F'], df_pred['Pts_pred_lr'], color='orange', label='Prediction: Linear Regression', marker='o')
plt.scatter(df['F'], df_pred['Pts_pred_dt'], color='green', label='Prediction: Decision Tree')
plt.legend();
```

```python
from sklearn.tree import plot_tree

plot_tree(model, feature_names=['F'], filled=True, rounded=True, fontsize=8, max_depth=2);
```
